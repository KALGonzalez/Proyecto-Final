Analisis sobre variables categoricas

```python
import matplotlib.pylot as plt
import seaborn as sns 
fig,axis = plt.subplots(2,3,figsize=(10,7))

sns.histplot(ax = axis[0, 0], data = df, x = "Transaction_Type").set(ylabel = None)
sns.histplot(ax = axis[0, 1], data = df, x = "Device_Type").set(ylabel = None)
sns.histplot(ax = axis[0, 2], data = df, x = "Location").set(ylabel = None)
sns.histplot(ax = axis[1, 0], data = df, x = "Merchant_Category")
sns.histplot(ax = axis[1, 1], data = df, x = "Card_Type")
sns.histplot(ax = axis[1, 2], data = df, x = "Authentication_Method")
             
plt.tight_layout()


plt.show()
```

Analisis sobre variables numericas

```python
fig, axis = plt.subplots(2, 3, figsize = (10, 7))

sns.regplot(ax = axis[0, 0], data = df, x = "Transaction_Amount", y = "Account_Balance")
sns.heatmap(df[["Account_Balance", "Transaction_Amount"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0], cbar = False)
sns.regplot(ax = axis[0, 1], data = df, x = "Transaction_Amount", y = "Avg_Transaction_Amount_7d").set(ylabel=None)
sns.heatmap(df[["Avg_Transaction_Amount_7d", "Transaction_Amount"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 1])
sns.regplot(ax = axis[0, 2], data = df, x = "Transaction_Distance", y = "Risk_Score").set(ylabel=None)
sns.heatmap(df[["Risk_Score", "Transaction_Distance"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 2])
# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()
```

```python
fig, axis = plt.subplots(2, 1, figsize = (10, 7))



sns.regplot(ax = axis[0], data = df, x = "Account_Balance", y = "Transaction_Amount")
# Correlation Matrix
sns.heatmap(df[["Transaction_Amount", "Account_Balance"]].corr(), annot = True, fmt = ".2f", ax = axis[1])

plt.tight_layout()
plt.show()
```
